# Xóa frontmatter (--- ... ---) trong file Markdown

Notebook đọc file .md, xóa toàn bộ khối nằm giữa dòng bắt đầu bằng `---` và dòng kết thúc bằng `---` (YAML frontmatter), rồi ghi lại nội dung đã xử lý.

## 1. Cấu hình

```python
from pathlib import Path

# Đường dẫn file .md cần xóa frontmatter (tương đối từ textbooks hoặc tuyệt đối)
file_path = Path("../Tiếng Anh 9/SGK Tiếng Anh 9 - Global Success/SGK Tiếng Anh 9 - Global Success.md")

# True = ghi đè file gốc; False = ghi ra file mới (thêm hậu tố _no_frontmatter)
overwrite_original = False
```

## 2. Đọc, xóa frontmatter, ghi file

```python
def _is_delimiter_line(line: str) -> bool:
    """True nếu dòng (sau khi strip) bắt đầu bằng '---' và kết thúc bằng '---'."""
    s = line.strip()
    return len(s) >= 3 and s.startswith("---") and s.endswith("---")


def remove_frontmatter(content: str) -> str:
    """1) Xóa khối frontmatter (từ dòng '---' đầu đến dòng '---' thứ hai).
    2) Xóa mọi dòng còn lại có dạng bắt đầu và kết thúc bằng '---' (vd: --- ... ---)."""
    lines = content.splitlines(keepends=True)
    result: list[str] = []
    in_frontmatter = False
    found_first = False
    for line in lines:
        stripped = line.strip()
        if stripped == "---":
            if not found_first:
                found_first = True
                in_frontmatter = True
                continue
            else:
                in_frontmatter = False
                continue
        if in_frontmatter:
            continue
        if _is_delimiter_line(line):
            continue
        result.append(line)
    return "".join(result)


# Resolve path: nếu tương đối thì tính từ thư mục chứa notebook (scripts)
if not file_path.is_absolute():
    file_path = Path.cwd() / file_path

if not file_path.exists():
    raise FileNotFoundError(f"Không tìm thấy file: {file_path}")

text = file_path.read_text(encoding="utf-8")
cleaned = remove_frontmatter(text)

if overwrite_original:
    out_path = file_path
else:
    out_path = file_path.parent / f"{file_path.stem}_no_frontmatter{file_path.suffix}"

out_path.write_text(cleaned, encoding="utf-8")
print(f"✓ Đã xóa frontmatter")
print(f"  - Đọc từ: {file_path}")
print(f"  - Ghi ra:  {out_path}")
```
